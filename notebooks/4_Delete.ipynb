## Delete the data in CDF

Let's take a quick look at how we can delete data from CDF. We will do this both to get an understanding of the deletion, but also to clean up the data we created in this course: 

### CogniteAsset

Listing all the CogniteAssets we created, that we should delete. 
If you have been following the course, you should have 8 CogniteAssets to delete

```python
from cognite.client.data_classes.filters import Prefix

asset_view_id = CogniteAsset.get_source()

assets_to_delete = client.data_modeling.instances.list(
    sources=asset_view_id,
    space="playground",
    filter=Prefix(("node", "externalId"), prefix),
)
assets_to_delete
```

```python
to_delete = assets_to_delete.as_ids()
res = client.data_modeling.instances.delete(nodes=to_delete)

print(f"Number of CogniteAssets deleted: {len(res.nodes)}")
```

### CogniteTimeSeries

Listing all the CogniteTimeSeries we created (should be 1), that we should delete:

```python
ts_view_id = CogniteTimeSeries.get_source()

ts_to_delete = client.data_modeling.instances.list(
    sources=ts_view_id,
    space="playground",
    filter=Prefix(("node", "externalId"), prefix),
)
ts_to_delete
```
